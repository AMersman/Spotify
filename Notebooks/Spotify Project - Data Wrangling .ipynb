```python
#import basic packages needed for data wrangling/cleaning
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
#import data for tracks features
df = pd.read_csv('../data/data.csv')
df.info()
```

```python
#for documentation regarding the tRack attributes see https://developer.spotify.com/documentation/web-api/reference/#/operations/get-audio-features
```

```python
#review first 10 rows of tracks using .head()
df.head(10)
```

```python
#remove unwanted characters from the artists column using str.replace()
df['artists'] = df['artists'].str.replace('[', '', regex=False)
df['artists'] = df['artists'].str.replace(']', '', regex=False)
df['artists'] = df['artists'].str.replace("'", '', regex=False)

#review changes to tracks['artists']
df.head()
```

```python
# Looking for missing values in the dataset
df.isna().sum()
```

```python
#check for duplicates
df[df['name']=='Landslide']
```

```python
#drop dupliactes
df.drop_duplicates(subset=['artists','name'],inplace = True)
```

```python
# Dropping values which are not required
df.drop(['id', 'release_date'], axis = 1, inplace = True)
```
